```python
import pandas as pd
import numpy as np
import subprocess
import time
```

```python
def creo_inputs(N):
    coord = []
    R = np.random.randint(1,10001)
    W = np.random.randint(1,N)
    V = np.random.randint(1,11)
    U = np.random.randint(1,V+1)
    for i in range (N):
        X = np.random.randint(0,1001)
        Y = np.random.randint(0,1001)
        coord.append(X)
        coord.append(Y)
    coords_str = ' '.join(str(num) for num in coord)
    input_str = f"1 {N} {R} {W} {U} {V} {coords_str}"
    return input_str

```

```python
inputs = []
tamaños = np.arange(100,10100,100)
for tam in tamaños:
    inputt = creo_inputs(tam)
    inputs.append(inputt)
```

```python
tiempos_ejec_dsu_opt = []
```

```python
tiempos_ejec_dsu_sin_opt = []
```

```python
# define the path to the executable
executable_path = r"C:/Users/Victoria/Downloads/algo3/tp2/ej3/ej3-dsu-opt"


# define a list of inputs
input_list = inputs

# define an empty list to store the execution times
execution_times = []

# run the executable for each input and measure the execution time
for i in range(len(inputs)):
    start_time = time.time()
    process = subprocess.Popen(executable_path, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, universal_newlines=True)
    stdout, stderr = process.communicate(input=inputs[i])
    end_time = time.time()
    execution_time = end_time - start_time
    tiempos_ejec_dsu_opt.append(execution_time)

```

```python
# define the path to the executable
executable_path = r"C:/Users/Victoria/Downloads/algo3/tp2/ej3/ej3-dsu-sin-opt"


# define a list of inputs
input_list = inputs

# define an empty list to store the execution times
execution_times = []

# run the executable for each input and measure the execution time
for i in range(len(inputs)):
    start_time = time.time()
    process = subprocess.Popen(executable_path, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, universal_newlines=True)
    stdout, stderr = process.communicate(input=inputs[i])
    end_time = time.time()
    execution_time = end_time - start_time
    tiempos_ejec_dsu_sin_opt.append(execution_time)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,10))
plt.scatter(tamaños,tiempos_ejec_dsu_opt,label = 'DSU optimizado')
plt.scatter(tamaños,tiempos_ejec_dsu_sin_opt,label = 'DSU sin optimizar')
plt.legend()
plt.yticks(fontsize =  14)
plt.ylabel('Tiempo',fontsize=15)
plt.xlabel('Cantidad de oficinas',fontsize = 15)
plt.title('Tiempo de ejecución en segundos por tamaño de entrada',fontsize=18)
plt.savefig('tiempo_ejec.png',bbox_inches='tight')
plt.show()
```
